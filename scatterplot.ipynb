```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(10, 90, size=(10))
```

```python
y = np.random.randint(1, 9, size=(10))
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y, marker='*', c='green', s=50)
```

```python
plt.scatter(x, y, marker='*', c='green', s=500)
```

```python
z = np.random.randint(1, 9, size=(10))
```

```python
plt.scatter(x, y, marker='*', c='green', s=80)
plt.scatter(x, z, marker='.', c='red', s=80)
```

```python
plt.scatter(x, y, marker='*', c='green', s=80)
plt.scatter(x, z, marker='.', c='red', s=80)
plt.legend(['xy', 'xz'])
```

```python
plt.subplot(1, 2, 1)
plt.scatter(x, y, marker='*', c='green', s=80)

plt.subplot(1, 2, 2)
plt.scatter(x, z, marker='.', c='red', s=80)
```

```python
plt.subplot(2, 1, 1)
plt.scatter(x, y, marker='*', c='green', s=80)

plt.subplot(2, 1, 2)
plt.scatter(x, z, marker='.', c='red', s=80)
```
