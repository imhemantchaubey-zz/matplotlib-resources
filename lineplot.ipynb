```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 11)
```

```python
x
```

```python
y = 2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y, x)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x--label")
plt.ylabel("y--label")
plt.show()
```

```python
plt.plot(x, y, color='r')
```

```python
plt.plot(x, y, color='violet', linestyle=':')
```

```python
plt.plot(x, y, color='violet', linestyle=':', linewidth=3)
```

```python
plt.plot(x,y, color='r', linewidth=3)
plt.title("Line Plot")
plt.xlabel("x--label")
plt.ylabel("y--label")
plt.show()
```

```python
z = (x+y)/2
```

```python
z
```

```python
plt.plot(x,y, color='r', linestyle=':', linewidth=3)
plt.plot(x,z, color='g', linestyle='-', linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
y = np.cos(x)
```

```python
y
```

```python
z
```
