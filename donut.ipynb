```python
import matplotlib.pyplot as plt
```

```python
sizes = [100, 500, 70, 54, 440]
```

```python
labels = ['Apple', 'Banana', 'Mango', 'Grapes', 'Orange']
```

```python
colors = ['#FF0000', '#0000FF', '#FFFF00', '#ADFF2F', '#FFA500']
```

```python
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
```

```python
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)
plt.show()
```

```python
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)
plt.pie([1], colors=['w'], radius=0.5)
plt.show()
```

```python
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', pctdistance=0.75, explode=explode)
plt.pie([1], colors=['w'], radius=0.75)
plt.show()
```
